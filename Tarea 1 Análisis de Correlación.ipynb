# Tarea en clase | Análisis de Correlación

```python
#Librerías
import pandas as pd
import numpy as np

#leer base de datos
df = pd.read_csv('CarPrice_Assignment.csv')
```

```python
#Delimitar las variables para el análisis
df = df[['price', 'carlength',  'carwidth',  'carheight',   'enginesize',  'boreratio',  'stroke', 'horsepower']]
```

```python
#imprimir matriz de correlación
df.corr()
cor = abs(df.corr())
cor
```
